```python
#3. Faktöriyel Hesaplama
def fac(a):
    if(a<0):
        return "the input cannot be negative !"
    elif(a==0):
        return 1
    else:
        return a*fac(a-1)
```

```python
#4. Listedeki Çift Sayıları Filtreleme
def filter_even_numbers(liste):
    odd_numbers=[]
    for i in liste:
        if(i%2==1):
            odd_numbers.append(i)
    for odd in odd_numbers:
        liste.remove(odd)
    return liste
```

```python
input4_1=[1, 2, 3, 4, 5, 6]
input4_2=[11, 23, 44]
print(filter_even_numbers(input4_1))
print(filter_even_numbers(input4_2))
```

```python
#5. İki Listenin Birleşimi ve Tekilleştirilmesi
def union_of_lists(liste1, liste2):
    set1= set(liste1)
    set2= set(liste2)
    set3= set1 | set2
    return list(set3)
```

```python
input5_1_1=[1, 2, 3]
input5_1_2=[3, 4, 5]
input5_2_1=[]
input5_2_2=[1, 1, 2]
print(union_of_lists(input5_1_1,input5_1_2))
print(union_of_lists(input5_2_1,input5_2_2))
```

```python
#6. Palindrom Kontrolü
def isPalindrom(string):
    string.replace(" ","")
    string=string.lower()
    length= len(string)
    if length==0:
        return "The given argument cannot be an empty string"
    for i in range(int(length/2)):
        if string[i]!=string[length-1-i]:
            return False
        return True
```

```python
input6_1="A man a plan a canal Panama"
input6_2="hello"
print(isPalindrom(input6_1))
print(isPalindrom(input6_2))
```

```python
#7. İki Sayının En Büyük Ortak Böleni (EBOB)
def ebob(int1, int2):
    if(int1<0 | int2<0):
        return "The given integers must be positive."
    min_n= min(int1,int2)
    ebob=0
    for i in range(min_n,0,-1):
        if((int1%i==0) and (int2%i==0)):
            ebob=i
            break
    return ebob
    
```

```python
input_7_1_1=54
input_7_1_2=24
input_7_2_1=7
input_7_2_2=13
print(ebob(input_7_1_1,input_7_1_2))
print(ebob(input_7_2_1,input_7_2_2))
```

```python
#8. Anagram Kontrolü
def check_anagram(str1, str2):
    str1.replace(" ","")
    str2.replace(" ","")
    str1= str1.lower()
    str2=str2.lower()
    check_flag=True
    for c in str1:
        if c not in str2:
            check_flag=False
    return check_flag
```

```python
input_8_1_1="listen"
input_8_1_2="silent"
input_8_2_1="hello"
input_8_2_2="world"
print(check_anagram(input_8_1_1, input_8_1_2))
print(check_anagram(input_8_2_1, input_8_2_2))
```

```python
#9. Listedeki Eksik Sayıyı Bulma
def find_missing_value(liste):
    '''assumed the given list always start from one based on given specifications'''
    length=len(liste)+1
    for i in range(1,length+1):
        if i not in liste:
            return i
        
```

```python
#10. Listeyi Belirli Bir Değere Göre Döndürme (Rotate)
def rotate_list(liste,k):
    length= len(liste)
    for i in range(k): # 0,1,..k-1
        last=liste[length-1]
        for j in range(1,length):
            liste[length-j]= liste[length-1-j]
        liste[0]=last
    return liste
            
```

```python
input10_1_1=[1, 2, 3, 4, 5]
input10_1_2=2
input10_2_1=[-1, -100, 3, 99]
input10_2_2=2
print(rotate_list(input10_1_1,input10_1_2))
print(rotate_list(input10_2_1,input10_2_2))
```

```python
#14. Maksimum Alt Dizi Toplamı
def sum_of_max_substring(liste):
    sum_=-456
    length= len(liste)
    for i in range(length):
        for j in range(1,length+1):
            if(sum(liste[i:j])>sum_):
                sum_=sum(liste[i:j])
    return sum_
```

```python
#15. Birleştirme Aralıkları (Merge Intervals)
def merge_intervals(intervals):
    if not intervals:
        return []
    # başlangıca göre sırala
    intervals.sort(key=lambda x: x[0])
    merged = [intervals[0]]

    for current in intervals[1:]:
        prev = merged[-1]
        if current[0] <= prev[1]: 
            prev[1] = max(prev[1], current[1])
        else:
            merged.append(current)

    return merged
```

```python
input15_1=[[1,3],[2,6],[8,10],[15,18]]
input15_2=[[1,4],[4,5]]
print(merge_intervals(input15_1))  
print(merge_intervals(input15_2))  
```
